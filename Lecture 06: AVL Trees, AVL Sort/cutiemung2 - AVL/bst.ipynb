```python
class BSTNode:
    def __init__(self, parent, key):
        self.key = key
        self.parent = parent
        self.left = None
        self.right = None
        self.subtree_size = 1
        self.height = None
    
    def update_sub_size(self):
        self.subtree_size = (0 if self.left is None else self.left.update_sub_size())\
            + (0 if self.right is None else self.right.update_sub_size())
        return self.subtree_size
    
    def reserve(self, t, k):
        if abs(self.key-t) < k:
            return None
        
        if t<self.key:
            if self.left is None:
                self.left = BSTNode(self, t)
                return self.left
            else:
                return self.left.reserve(t,k)
        else:
            if self.right is None:
                self.right = BSTNode(self,t)
                return self.right
            else:
                return self.right.reserve(t,k)
        
        
    def insert(self, key):
        self.subtree_size += 1
        if key<=self.key:
            if self.left is None:
                self.left = BSTNode(self, key)
                return self.left
            else:
                return self.left.insert(key)
        else:
            if self.right is None:
                self.right = BSTNode(self,key)
                return self.right
            else:
                return self.right.insert(key)
            
    def find(self, key):
        if self.key==key:
            return self
        elif key<self.key:
            if self.left is None:
                return None
            else:
                return self.left.find(key)
        else:
            if self.right is None:
                return None
            else:
                return self.right.find(key)
                
    def rank(self, key):
            left_subtree_size = 0 if self.left is None else self.left.subtree_size
            if key<self.key:
                if self.left is None:
                    return 0
                else:
                    return self.left.rank(key)
            elif key>self.key:
                if self.right==None:
                    return left_subtree_size+1
                else:
                    return self.left.subtree_size+1+self.right.rank(key)
            else: 
                return left_subtree_size+1
                
        
```

```python
class BST:
    def __init__(self):
        self.root = None
        
    def reserve(self, t, k):
        if self.root == None:
            self.root = BSTNode(None, t)
        else:
            retval = self.root.reserve(t,k)
            self.root.update_sub_size()
            return retval
    
    def insert(self, key):
        if self.root is None:
            self.root = BSTNode(None, key)
            return self.root
        else:
            return self.root.insert(key)
            
    def find(self, key):
        if self.root is None:
            return None
        else:
            return self.root.find(key)
            
    def rank(self, key):
        if self.root is None:
            return None
        else:
            return self.root.rank(key)
        
    def __str__(self):
            if self.root is None: return '<empty tree>'
            def recurse(node):
                if node is None: return [], 0, 0
                label = str(node.key)
                left_lines, left_pos, left_width = recurse(node.left)
                right_lines, right_pos, right_width = recurse(node.right)
                middle = max(right_pos + left_width - left_pos + 1, len(label), 2)
                pos = left_pos + middle // 2
                width = left_pos + middle + right_width - right_pos
                while len(left_lines) < len(right_lines):
                    left_lines.append(' ' * left_width)
                while len(right_lines) < len(left_lines):
                    right_lines.append(' ' * right_width)
                if (middle - len(label)) % 2 == 1 and node.parent is not None and \
                   node is node.parent.left and len(label) < middle:
                    label += '.'
                label = label.center(middle, '.')
                if label[0] == '.': label = ' ' + label[1:]
                if label[-1] == '.': label = label[:-1] + ' '
                lines = [' ' * left_pos + label + ' ' * (right_width - right_pos),
                         ' ' * left_pos + '/' + ' ' * (middle-2) +
                         '\\' + ' ' * (right_width - right_pos)] + \
                  [left_line + ' ' * (width - left_width - right_width) +
                   right_line
                   for left_line, right_line in zip(left_lines, right_lines)]
                return lines, pos, width
            return '\n'.join(recurse(self.root) [0])
```
